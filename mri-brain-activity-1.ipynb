```python
# Plotting original and squared values
plt.figure(figsize=(8, 5))

for key, value in seizure_vote.items():
    plt.plot(value, label=key)

plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Original Values')
plt.legend()
plt.show()

# Applying log transformation to seizure data
log_seizure_vote = {}

for key, value in seizure_vote.items():
    log_seizure_vote[key] = np.log1p(value)

# Plotting log transformed values
plt.figure(figsize=(8, 5))

for key, value in log_seizure_vote.items():
    plt.plot(value, label=key)

plt.xlabel('Index')
plt.ylabel('Log Transformed Values')
plt.title('Log Transformed Values')
plt.legend()
plt.show()
```
